# Mutable vs. Immutable Objects

* **Immutable** objects CANNOT be changed in place after they are created.
    * The variable name can be reused but a reassignment will store the value in a new memory location
* **Mutable** objects CAN be changed in place after they are created.
    * Be careful because other variables stored in the same memory location can change the value without you realizing!
        * Making a 'copy' that assigns the value in a new memory location will avoid this

# Properties of Objects

Remember! All objects have a value, type, AND memory location (ID)

Check for equality using **double equal signs** 

A ***single* equal sign** is an **assignment** (Used when you are defining a new variable)

A ***double* equal sign** is a **question** that asks "Are these things equal?"

How do we check if the **memory location** is equal to the location where we think it lives?

If we want to test if the **value** of **two variables** are the same

If we want ot test if the **memory location** of **two variables** are the same

OR! we can do the same thing using the keyword **is**

**is** asks "Are these things stored in the same memory location?"

**is** does *NOT* ask about value equality!

# Immutable Objects

## Integers are immutable

When you reassign an integer it gets stored in a new memory location

Even if you reassign it with the same value!

* **Immutable** objects CANNOT be changed after they are created.
    * The variable name can be reused but a reassignment will store the value in a new memory location

## Strings are immutable

When you reassign a string it gets stored in a new memory location

Even if you reassign it with the same value!


Two different strings with the same value... 

Do not have the same memory location if they are assigned their values independently

Assigning a new string to an old string means they will have the same value AND memory location

As soon as you reassign one of the variables (even if you reassign it with the same value), 

they are no longer in the same memory location.

Remember how to index a sequence to extract an element from the container

Can you reassign just that one element, but leave the rest of the string the same?

What would be the memory address of the string if you do this?

* **Immutable** objects CANNOT be changed in place after they are created.
    * The variable name can be reused but a reassignment will store the value in a new memory location

## Tuples are immutable

Remember how to make a tuple

The comma is what is most important!

Side Note: Accidentally inserting commas into your code often makes accidental tuples 

Tuples are also immutable, so every time you reassign them, they get a new memory location

Tuples are also sequences, so you can index them!

Can you reassign just that one element, but leave the rest of the tuple the same?

What would be the memory address of the tuple if you do this?

* **Immutable** objects CANNOT be changed in place after they are created.
    * The variable name can be reused but a reassignment will store the value in a new memory location

# Mutable Objects!

## Lists are Mutable!

If you reassign the whole list it will store it in a new memory location (same as immutable)

BUT if you just reassign a part of the list the memory location will stay the same!

* **Mutable** objects CAN be changed in place after they are created.

BEWARE!! :

What if you don't want *n* to change after you assign it?

Make a 'copy' of *m* and store it in *n*!

The main difference between a list and tuple is

* Lists are Mutable
* Tuples are **Im**mutable

Lists are more common, but sometimes tuples are used to avoid accidents of the BEWARE case above
