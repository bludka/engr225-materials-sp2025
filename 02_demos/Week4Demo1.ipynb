## Indentation Matters in Python!!

The branch you are on, is determined by indendation, so it is very important!!!

It is best practice to use 4 spaces to perform an indent

You might be able to get away with 2 spaces, or the tab button, but it is not recommended.

You will typically get an error if you try to mix and match indentation methods.
