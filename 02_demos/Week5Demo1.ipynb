# While Loops

While loops are not used very often. 

They should ONLY be used if there is no way to know how many times the loop will run before you enter the loop.

Students typically have an easier time understanding while loops than for loops, so we cover while loops first.

While loop syntax:

```
while <condition>:
    <actions to take each time through loop>
    #loop only executes when condition is true
    #loop finishes when condition is false
```

While loops are prone to accidental infinite loops!

One example of when you need a while loop, is when the user is asked for input on each loop iteration

If we want to keep track of what happens during each loop, 

we should create a **storage variable** before the loop.

What if instead we wanted to keep track of the number of orders that were placed?

For this let's create a **counting variable** before the loop.

Another time we might use a while loop is when we are dealing with a random number in order to decide whether to continue a repetition.

# For loops

# Most loops are... for loops!  

# Use when you need to do something repeatedly for a *known number of times*.

For loop syntax:



```
for <loop variable> in <container>:
    <actions to take each time through loop>
    #loop variable changes to next value in container with each iteration
    #loop finishes when loop variable has used all values in the container

```

Since the size of the container is known before the loop is entered, we know the number of times the loop will execute.
