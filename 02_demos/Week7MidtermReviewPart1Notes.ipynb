# Properties of an object:
* value - what is stored in object
* type - how the object is stored
* id - where the object is stored

```python
a = 5.
print(a) #value
print(type(a)) #type
print(id(a)) #id
```

`is` checks if two variables are stored in the same place. THIS IS NOT THE SAME AS `==` !!!

```python
a = 1000
b = 10**3
print(a==b)#even though a and b are equal
print(a is b) #they are not stored in the same place
print(id(a)) #address where a is stored
print(id(b)) #address where b is stored
```

Difference between a list and a tuple:
You can not reassign PART of a tuple.

```python
a = [1,2,3,4]
b = (1,2,3,4)
a[0] = 'hi' #allowed for lists
print(a)
b[0] = 'hi' #not allowed for tuples
print(b) #tuples are used when you want to be extra careful that part of your data will not be reassigned.
```

```python
b = (5,6,7,8) #but you can still change the tuple if you reassign it completely
```

# While loops

Only use a while loop if you don't know how many times the loop will execute.

```python
go = 'Yes'
count = 1
while go == 'Yes':
    print(count)
    go = input('Again?: ')
    count = count + 1
```

# Indexing, Slicing, Striding

```python
a = 1,2,3,4
print(type(a))
```

Indexing

```python
a = 1,2,3,4
b = a[0]
print(b)
```

```python
a = 1,2,3,4
c = a[-1]
print(c)
```

Slicing

```python
a = 1,2,3,4
d= a[0:2]
print(d)
```

```python
a = 1,2,3,4
e = a[:2]
print(e)
```

```python
a = 1,2,3,4
f = a[2:-1]
print(f)
```

```python
a = 1,2,3,4
g = a[2:]
print(g)
```

Striding

```python
a = 1,2,3,4
h = a[0:4:2] #every other
print(h)
```

```python
a = 1,2,3,4
m = a[::2] #every other
print(m)
```

```python
a = 1,2,3,4
n = a[::-1] #backwards!
print(n)
```

Indexing nested containers

```python
a = [[1,2,3],
     [4,5,6]]

print(a[0])
print(a[0][1])
print(a[-1][-1])
```
