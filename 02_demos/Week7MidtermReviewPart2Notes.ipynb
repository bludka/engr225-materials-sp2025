# For loops

Looping over a value

```python
num = 5,6,7,8
for n in num: #n is the loop variable and the tuple num is the container
    print(n)
```

```python
fruits = ['apple','pear','lime','lemon','dragon fruit']
for f in fruits: #f is the loop variable and the list fruits is the container
    print(f)
```

```python
a = 'apple'
for char in a: #char is the loop variable and the string a is the container
    print(char)
```

Looping over an index

(Useful if you want to extract multiple values in a loop. For example below I want to extract a value from `student1_grades` AND a value from `student2_grades` in order to be able to calculate an average for a test)

```python
# Calculate the average of the students grades on each test
student1_grades = [100,100,98,92] #This student got 100 on the first test, 100 on the second test, 98 on the 3rd test..
student2_grades = [89,90,72,92] #This student got 89 on the first test, 90 on the second test, 72 on the 3rd test..
for i in range(len(student1_grades)): #len(student_grades)=4, so range(4) goes from 0 to 3
    #print(i)
    avg = (student1_grades[i] + student2_grades[i])/2
    print(avg)
# Check - does your answer make sense? Are the average grades what you expect?
```

Appending a list in a loop

```python
#Extract all positive numbers from my_list and store them in a new list
my_list = [4, -6, 19, 20, -99, 57]
my_pos_list = []
for i in range(len(my_list)):
    if my_list[i]>0:
        my_pos_list.append(my_list[i])
print(my_pos_list)
```

Nested loops

```python
#imagine you want to print out tickets for each seat in an auditorium.

rows = ['A', 'B', 'C', 'D']
seatnums = [1,2,3,4,5,6] #each row has this many seats

for r in rows:
    for s in seatnums:
        print(r,s)
```

```python
#what if we wanted to store all the tickets in a list
rows = ['A', 'B', 'C', 'D']
seatnums = [1,2,3,4,5,6] #each row has this many seats

ticket = []
for r in rows:
    for s in seatnums:
        ticket.append(r+str(s))
print(ticket)
```

```python
#now let's print out all the items in ticket
for t in ticket:
    print(t)
```

```python
#Or maybe we want to store the row and seat as separate elements in a list
rows = ['A', 'B', 'C', 'D']
seatnums = [1,2,3,4,5,6] #each row has this many seats

tickets = []
for r in rows:
    row_tickets =[]
    for s in seatnums:
        row_tickets.append([r,s])
    tickets.append(row_tickets)
```

```python
#now let's print out all the items in tickets
for i in range(len(tickets)):
    for j in range(len(tickets[0])):
        print(tickets[i][j])
```
