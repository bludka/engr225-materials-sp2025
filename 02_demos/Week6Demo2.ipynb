## Nested Summations

$$ \sum_{i=1}^{2} \sum_{j=1}^{2} ({i*j}) $$

$$ \sum_{i=1}^{2} ({i*1 + i*2}) $$
$$ \sum_{i=1}^{2} ({3*i}) $$
$$ (3*1) + (3*2) = 3+6=9 $$

OR you can think of it as:
$$ \sum_{i=1}^{2} \sum_{j=1}^{2} ({i*j}) $$

i = 1, j = 1 --> 1*1 = 1

i = 1, j = 2 --> 1*2 = 2

i = 2, j = 1 --> 2*1 = 2

i = 2, j = 2 --> 2*2 = 4

1+2+2+4 = 9
